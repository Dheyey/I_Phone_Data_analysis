```python
import pandas as pd 
```

```python
#Read the iphone.csv dataset to pandas dataframe
idata = pd.read_csv('iphone.csv')
```

```python
#1 - The column names have spaces . rename the column names to have underscore '_' instead of space (try to do in one go instead of specifying each column nam in rename method)

idata.columns = idata.columns.str.replace(' ', '_')
```

```python
idata['Number_Of_Ratings'] = idata['Number_Of_Ratings'].fillna(idata['Number_Of_Ratings'].mean())
```

```python
idata['Star_Rating'] = idata['Star_Rating'].fillna(idata.groupby('Ram')['Star_Rating'].transform('mean'))
```

```python
idata['Discount_Percentage'] = ((idata['Mrp']-idata['Sale_Price'])/idata['Mrp']) * 100
idata
```

```python
def extract_color(product_name):
    # Split the product name by '(' and ')' to get parts within parentheses
    parts = product_name.split('(')
    if len(parts) > 1:
        # Check if the second part contains a comma (,) to identify color
        color_part = parts[1].strip()
        if ',' in color_part:
            color = color_part.split(',')[0].strip()
            return color
    return None
```

```python
idata['Color'] = idata['Product_Name'].apply(extract_color)
```

```python
idata
```

```python
def extract_c(Produ):
    parts = Produ.split('(')
    right = parts[1].split(',')
    color = right[0]
    return color
```

```python
idata['Color1'] = idata['Product_Name'].apply(extract_c)
```

```python
def extract_model(product_name):
    parts = product_name.split(' ')
    model_name = ' '.join(parts[1:3])
    return model_name
```

```python
idata['model_name'] = idata['Product_Name'].apply(extract_model)
```
